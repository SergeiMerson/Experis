```python
list_of_tables = [
    {'name': 'qCategories',
     'columns': [('CategoryID', 'SMALLINT', ' PRIMARY KEY'),
                 ('Category', 'VARCHAR(60)', 'NOT NULL')],
     'values': [(0, 'Any Category')]},

    {'name': 'qDifficulties',
     'columns': [('DifficultyID', 'VARCHAR(10)', 'PRIMARY KEY'),
                 ('Difficulty', 'VARCHAR(20)', 'NOT NULL')],
     'values': [('0', 'Any Difficulty'),
                ('easy', 'Easy'),
                ('medium', 'Medium'),
                ('hard', 'Hard')]},

    {'name': 'qTypes',
     'columns': [('TypeID', 'VARCHAR(10)', 'PRIMARY KEY'),
                 ('Type', 'VARCHAR(20)', 'NOT NULL')],
     'values': [('0', 'Mixed Types'),
                ('multiple', 'Multiple Choice'),
                ('boolean', 'True/False')]},

    {'name': 'Questions',
     'columns': [('QuestionID', 'INT', 'IDENTITY PRIMARY KEY'),
                 ('CategoryID', 'SMALLINT', 'FOREIGN KEY REFERENCES qCategories(CategoryID)'),
                 ('DifficultyID', 'VARCHAR(10)', 'FOREIGN KEY REFERENCES qDifficulties(DifficultyID)'),
                 ('TypeID', 'VARCHAR(10)', 'FOREIGN KEY REFERENCES qTypes(TypeID)'),
                 ('Question', 'VARCHAR(300)', 'NOT NULL UNIQUE')],
     'values': []},

    {'name': 'qAnswers',
     'columns': [('AnswerID', 'INT', 'IDENTITY PRIMARY KEY'),
                 ('QuestionID', 'INT', 'NOT NULL FOREIGN KEY REFERENCES Questions(QuestionID)'),
                 ('Answer', 'VARCHAR(99)', 'NOT NULL'),
                 ('IsCorrect', 'BIT', 'NOT NULL DEFAULT 0')],
     'values': []},

    {'name': 'Users',
     'columns': [('UserID', 'INT', 'IDENTITY PRIMARY KEY'),
                 ('UserName', 'VARCHAR(30)', 'NOT NULL')],
     'values': []},

    {'name': 'Games',
     'columns': [('GameID', 'INT', 'IDENTITY PRIMARY KEY'),
                 ('UserID', 'INT', 'FOREIGN KEY REFERENCES Users(UserID)'),
                 ('GameDate', 'DATE', 'NOT NULL CHECK (GameDate <= GETDATE())')],
     'values': []},

    {'name': 'GameResults',
     'columns': [('GameID', 'INT', 'FOREIGN KEY REFERENCES Games(GameID)'),
                 ('QuestionID', 'INT', 'FOREIGN KEY REFERENCES Questions(QuestionID)'),
                 ('AnswerID', 'INT', 'FOREIGN KEY REFERENCES qAnswers(AnswerID)')],
     'values': []},
]
```

```python
database.initialize_basic_tables(list_of_tables)
```
